```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import confusion_matrix, accuracy_score
```

```python
# 1. Load the dataset
url = "http://archive.ics.uci.edu/ml/machine-learning-databases/wine-quality/winequality-white.csv"
data = pd.read_csv(url, sep=';')
```

```python
# 2. Check for missing values
missing_values = data.isnull().sum()
print("Missing values in the dataset:")
print(missing_values)
```

```python
# 3. Compute correlation matrix
corr_matrix = data.drop('quality', axis=1).corr()

corr_matrix
```

```python
# 4. Splitting the dataset into Independent and Dependent Features
X = data.drop('quality', axis=1)
y = data['quality']
```

```python
# 5. Normalize the dataset
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
# 6. Split the dataset into training and test sets
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)
```

```python
# 7. Check whether the target variable is balanced or not
target_balance = y.value_counts(normalize=True)
print("Target Variable Balance:")
print(target_balance)
```

```python
# 8. Decision Tree algorithm
dt_classifier = DecisionTreeClassifier(random_state=42)
dt_classifier.fit(X_train, y_train)
```

```python
# 9. Confusion Matrix
y_pred = dt_classifier.predict(X_test)
conf_matrix = confusion_matrix(y_test, y_pred)
print("Confusion Matrix:")
print(conf_matrix)
```

```python
# 10. Model Accuracy
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
```
